# 2. Pre-processing

```python
# Data Understanding

df.info()
df.describe()
```

```python
# Penghapusan Fitur Tidak Relevan

colDrop = ['Product ID', 'Product Name', 'Supplier ID']

df.drop(colDrop, axis=1, inplace=True)
df.head()
```

```python
# Encoding Fitur Kategorikal

from sklearn.preprocessing import OrdinalEncoder
categoryCol = [
    'Category',
    'Customer Gender',
    'Customer Location',
    'Shipping Method',
    'Seasonality'
]

age_cat = [['18-24', '25-34', '35-44', '45-54', '55+']]
ordinalEncoder = OrdinalEncoder(categories=age_cat)

df['Customer Age Group'] = ordinalEncoder.fit_transform(df[['Customer Age Group']])
df = pd.get_dummies(df, columns=categoryCol, drop_first=True)

df.head()
```

```python
# Feature Engineering

threshold = df['Return Rate'].median()
df['Returned'] = df['Return Rate'].apply(lambda x: 1 if x > threshold else 0)

df.drop('Return Rate', axis=1, inplace=True)
df.head()
```

```python
# Feature Scaling

from sklearn.preprocessing import StandardScaler

fiturNumeric = ['Price', 'Discount', 'Tax Rate', 'Stock Level', 'Shipping Cost', 'Popularity Index']
scaler = StandardScaler()
df[fiturNumeric] = scaler.fit_transform(df[fiturNumeric])
df.head()
```
